```python
df = pd.read_csv('/content/1) iris.csv')
df.head()
```

```python
df.info()
```

```python
print('Number of duplicated rows:', df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
print(df.isnull().sum())
```

```python
df.describe()
```

```python
species_counts = df['species'].value_counts()
print("\nDistribution of Species:\n", species_counts)
```

```python
pd.plotting.hist_frame(df, bins=20, figsize=(10, 6))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='petal_length', data=df)
plt.title('Box Plot of Petal Length by Species')
plt.show()
```

```python
# Scatter plot
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.show()
```

```python
correlation_matrix = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr() # Select only numeric columns for correlation
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Iris Dataset')
plt.show()
```

```python
#barplot for species distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='species', data=df, palette='viridis')
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Species Distribution')
plt.savefig('species_distribution.png')
plt.show()

#scatterplot for correlation sepal_length & sepal_width
plt.figure(figsize=(6, 4))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Species')
plt.title('Sepal Length vs Sepal Width')
plt.savefig('sepal_scatterplot.png')
plt.show()

#lineplot for sepal_length per sample
plt.figure(figsize=(6, 4))
plt.plot(df.index, df['sepal_length'], color='green', marker='o', linestyle='-')
plt.title('Sepal Length per Sample')
plt.xlabel('Index')
plt.ylabel('Sepal Length')
plt.grid(True)
plt.tight_layout()
plt.savefig('sepal_lineplot.png')
plt.show()
```

```python
sns.pairplot(df, hue='species', palette='viridis', diag_kind='kde')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)  # Adjust title position
plt.show()
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=df, kind='kde', hue='species', palette='viridis')
plt.suptitle('Joint Plot of Sepal Length vs. Sepal Width', y=1.02)
plt.show()
```

```python
from pandas.plotting import radviz

plt.figure(figsize=(8, 6))
radviz(df, 'species', colormap='viridis')
plt.title('Radviz Plot of Iris Dataset')
plt.show()
```
